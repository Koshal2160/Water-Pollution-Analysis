```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/water_pollution.csv")

df.shape
df.info()
df.head(10)
df.isna().sum()          # check missing values
df.duplicated().sum()    # check duplicates
```

```python
df = df.drop_duplicates()
df.shape
```

```python
text_cols = ["State", "City", "Water Source", "Pollutant Type", "Pollution Level", "Safe Limit"]

for col in text_cols:
    df[col] = df[col].astype(str).str.strip()          # remove stray spaces
    df[col] = df[col].str.title()                       # consistent casing

df["Safe Limit"] = df["Safe Limit"].str.capitalize()     # Yes/No
```

```python
df["Year"] = df["Year"].astype(str).str.replace(".0", "", regex=False)
df["Year"] = pd.to_numeric(df["Year"], errors="coerce").astype("Int64")
```

```python
# Numeric columns: fill with column median
num_cols = ["pH", "Dissolved Oxygen", "Temperature"]
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

# Categorical columns: fill with mode or "Unknown"
for col in text_cols + ["Year"]:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df["pH"] = df["pH"].round(2)
df["Dissolved Oxygen"] = df["Dissolved Oxygen"].round(2)
df["Temperature"] = df["Temperature"].round(1)
df["Year"] = df["Year"].astype(int)
```
